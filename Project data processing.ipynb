```python
import pandas as pd # import pandas library
import numpy as np # import numpy library
import matplotlib.pyplot as plt # imports the matplotlib library
```

```python
df1 = pd.read_csv('movies_metadata.csv' ,na_values=["?","n.a", "_"], sep=',') # import the dataset and for uncleaned dataset in some
# cases we use different sep='\t .... etc' instead of sep = ','
# na_values=["?","n.a", "_"] it is used for changing the values of the dataset
# For example if in our dataset we may have ????____ or n.a instead of NaN we replace them with NaN
```

### Now we have to check if our dataset is cleaned so we call it first and see in it

```python
df1 # We see that we have NaN values we also see that inside that row we have useful contents
# In some situation we may not want a row if there Nan values and we drop them to do that
# We use df.dropna()  this command will erase all the rows that contains NaN values
# On the other hand if we still need whats inside we fill them with a value for example with zero value. To do that and have it for later we set a variable for that

df = df1.fillna(0)
df 

# But we see that we have strings instead of numbers in budget column
```

```python
letters = 'qwertyuiopasdfghjklzxcvbnm' 
all_letters = letters.upper() + letters

# Here we construct a function which it will get every element of the column and test it if that element is a str
# if in the row the def will detect a string than it will return a NaN value

def replace_element(element):
    for char in str(element):
        if char in all_letters:
            return np.nan
    else:
        return element
    
df.budget = df.budget.apply(replace_element)
        
    
```

```python
df = df.fillna(0)  # But its not on our benefit to keep a NaN value so we replace it with a '0'
```

# Creating a pie chart to show the low , medium and high budget movies

```python
max_budget = df.budget.astype(int).max()
min_budget = df.budget.astype(int).min()
segment_size = (max_budget - min_budget)/3  # vlera e mesit

s0 = min_budget
s1 = min_budget + segment_size
s2 = min_budget + 2*segment_size
s3 = max_budget

# s0|------s1|-------s2|--------s3|

for index, row in df.iterrows():
    if s0 <= int(row.budget) < s1:
        df.loc[index,'budget_level']  = 'low_budget'
    elif s1 <= int(row.budget) <s2:
        df.loc[index, 'budget_level']  = 'medium_budget'
    else:
        df.loc[index, 'budget_level']  = 'high_budget'


plt.pie(df.budget_level.value_counts(), labels= df.budget_level.value_counts().keys(), autopct='%.2f%%')
plt.legend(bbox_to_anchor=(-0.1,0.5) )
plt.show()
```

# Creating a bar chart , finding for each film how many languages are spoken 

```python
df['nr_languages'] = pd.Series() # creates a dataframe series
df['nr_languages'] = df.spoken_languages.str.count('name') # finds how many time the word name was used in each row of the spoken language columns and puts it in the nr_languages column

# Creates the bar chart
plt.bar(df.nr_languages.value_counts().keys(), df.nr_languages.value_counts())
plt.xlabel('Nr. of spoken languages')
plt.ylabel('Amount Of films')
plt.show()
```

# Finding how many movies there are for each type of genre 

```python
def get_from_list(element):
    element = eval(element)
    list_of_genres = []
    for d in element:
        list_of_genres.append(d['name'])
    return list_of_genres
        
df.genres = df.genres.apply(get_from_list)
#df.production_companies = df.production_companies.apply(get_from_list)
```

```python
genres_dictionary = {}

for list_of_genres in df.genres:
    for element in list_of_genres:
        if element in genres_dictionary.keys():
            genres_dictionary[element] += 1
        else:
            genres_dictionary[element] = 1
genres_dictionary
```
